```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model_path = r"D:\LLM_Demos\models\distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path, torch_dtype="auto")

tokenizer.pad_token = tokenizer.eos_token

input_text = "Once upon a time"
input_ids = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True).input_ids

output = model.generate(
    input_ids,
    max_length=100,  # Increase length for more meaningful completion
    temperature=0.8,  # Introduce randomness (lower = more predictable, higher = more creative)
    top_k=50,  # Limit vocabulary to top 50 tokens at each step (prevents random nonsense)
    top_p=0.9,  # Use nucleus sampling for better diversity
    repetition_penalty=1.2,  # Penalize repeating phrases
    pad_token_id=tokenizer.eos_token_id,
    attention_mask=torch.ones_like(input_ids)  # Ensure proper token attention
)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```
